```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/DATASETS/Iris.csv")
df
```

```python
df['Species'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Species']=le.fit_transform(df['Species'])
df.dtypes
```

```python
df['Species'].value_counts()
```

```python
X=df.iloc[:,1:-1]
X
```

```python
y=df.iloc[:,-1].values   #array akyvakkn ipo 1D aan
y
```

0->[1 0 0],,1->[0 1 0], 2->[0 0 1]....next ingna aknm  y na

```python
from tensorflow.keras.utils import to_categorical
y_new=to_categorical(y)
y_new
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X=scaler.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y_new,test_size=0.3,random_state=14)
```

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
ann=Sequential()
ann.add(Dense(units=4,activation='relu'))
ann.add(Dense(units=5,activation='relu'))
ann.add(Dense(units=6,activation='relu'))
ann.add(Dense(units=3,activation='softmax'))   #3neurons coz 3 ootputs indllo...if bin class we can use softmax..but we shpuld give neurons=2 and y shoukd be changed like abouve
```

```python
ann.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(X_train,y_train,batch_size=10,epochs=100)
```

```python
y_new=ann.predict(scaler.transform([[5.1,	3.5,	1.4, 0.2]]))
print(y_new)
ind=y_new.argmax()
print(ind)
result=le.inverse_transform([ind])   #inverse_transform method to map the index ind back to the original class label using a label encoder (le
result
```

```python
y_pred=ann.predict(X_test)
y_pred=y_pred.argmax(axis=1)
y_pred
```

```python

y_test=y_test.argmax(axis=1)  #numpy array row axis=1
y_test
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
